# Automated scoring

```python
# Script for defining the male and female distribution in the crawled set of images
# Uses api from sightengine to detect the gender
# Uses another api from haystack.ai to get the automated score. 

import random
import os
import json
import requests
import csv
from sightengine.client import SightengineClient

imgs = os.listdir("../data")
maleSpeakers = []
femaleSpeaker = []
finalImgList = []

# two profiles created at sighengine to request the crawled images 
# the api has limitation on requests per user.
sightengineKeys = [
    {'user': '1587401878', 'password': 'WNRqW34htp5D5eKBXUxU', 'from': 0, 'until': int(len(imgs) /2) },
    {'user': '1784535781', 'password': '9pt7WtpbMDaYZimYnVcq', 'from': int(len(imgs) /2), 'until': len(imgs) } ]


# detecting gender for each of the picture downloaded by using two keys of the sightengine api.
for key in range (0, len(sightengineKeys)):
    client = SightengineClient(sightengineKeys[key]["user"], sightengineKeys[key]["password"])
    for i in range(sightengineKeys[key]["from"], sightengineKeys[key]["until"]):
        print("detecting gender...", i, "/", len(imgs), imgs[i])
        try:
            output = client.check('face-attributes').set_file("../data/" + imgs[i])
            femalescore = output["faces"][0]["attributes"]["female"]
            malescore = output["faces"][0]["attributes"]["male"]
        except:
            print("something went wrong")  
            continue
        if(femalescore >  malescore): femaleSpeaker.append(imgs[i])
        else: maleSpeakers.append(imgs[i]) 

```

```python
# defining random subset of the picture with the distribution of male and female proportional of the total set of the images

myRandomMales = random.sample(range(0, len(maleSpeakers)), 60)
myRandomFemales = random.sample(range(0, len(femaleSpeaker)), 40)

finalImgList = []

for number in myRandomMales:
    finalImgList.append(maleSpeakers[number])

    
for number in myRandomFemales:
    finalImgList.append(femaleSpeaker[number])
```

```python
# haystack api limits the number of requests per user. thats why are created more profiles and 100 pictures are splitted by requesting 8pictures per one key

haystackKeys = [
        {'key': '442c89077a0c00ab1d8bc0ea111c4cfc', 'from': 0,  'until': 7},
        {'key': 'f50d7f1e2453e31e84323c61d46024bb', 'from': 7, 'until': 14},
        {'key': '2244077d00b8676da32a0e4d904c4bff', 'from': 14, 'until': 21},
        {'key': 'adcaa3a29cef213acced9a31998c4619', 'from': 21, 'until': 28},
        {'key': 'b188d7ec6c7bc72c0decf772d36e387a', 'from': 28, 'until': 35},
        {'key': '93bc17ca2d4e43dd4849c756805b6f5d', 'from': 35, 'until': 42},
        {'key': '9a48910789f1cf2057f1afd7f4350946', 'from': 42, 'until': 49},
        {'key': '12f97da1ce45303cd932c202a9f10400', 'from': 49, 'until': 56},
        {'key': '82a55edfb1606a3bbeaa88b3bf9fed36', 'from': 56, 'until': 63},
        {'key': '705eeaaab79b6051c4bde417a8fe63f6', 'from': 63, 'until': 70},
        {'key': '35810a809ff7f5b4048d4437319fe90d', 'from': 70, 'until': 77},
        {'key': 'd98aed9cf9b1ac04e1a1377afd7ac685', 'from': 77, 'until': 84},
        {'key': 'bf86c9027f0f000b329d8b03feafade0', 'from': 84, 'until': 91},
        {'key': '4f57c9c4cadb8cf1f00999a993cc044c', 'from': 91, 'until': 100}]

scored = []

for key in range(0, len(haystackKeys)):
    for i in range(haystackKeys[key]["from"], haystackKeys[key]["until"]):
        try:
            print("scoring...", i, "/", len(finalImgList), finalImgList[i])
            r = requests.post( "https://api.haystack.ai/api/image/analyze?output=json&apikey=" + haystackKeys[key]["key"] ,data=open("../data/" + finalImgList[i], 'rb'))
            json_data = json.loads(r.text)
            if(json_data['people']):
                if finalImgList[i] in femaleSpeaker:
                    gender = 'f'
                else: gender = 'm'
                scored.append(type('',(object,),{'img':finalImgList[i],'score':json_data['people'][0]['attractiveness'],'gender':gender})())
        except OSError as err:
            print("ERROR", format(err))  
```

```python
# creating csv file where the pictures are stored with the detected gender and the attractivness score provided by the api
with open('../output/scored-images.csv', 'w') as csvfile:
    fieldnames = ['Image','Gender', 'API score', 'Human rating']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in range(0, len(scored)):
        writer.writerow({ 'Image': scored[row].img, 'Gender': scored[row].gender, 'API score': scored[row].score, 'Human rating':'' })
```
