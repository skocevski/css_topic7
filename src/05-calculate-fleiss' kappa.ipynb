```python
#This is a program to calculate the inter rater agreement using Fleiss' Kappa
#Explanation of the formula and other information regarding Fleiss' Kappa is mentioned in detail in the report
#Chiranth

import xlrd
from collections import Counter

xl = xlrd.open_workbook('../output/kappa_calculation.xlsx')

#Read the xls sheet into the object
first_sheet = xl.sheet_by_index(0)

#Map to store speaker ID and its raters count. 
#Map key is speaker ID and map value is rater count. 
spker_map = dict()
rating_matrix = []
#Find the speaker ID in the map. 
#if present, increment its number of raters by 1
#if not present, initialize the number of raters to 1
def updateNumOfRaterToSpeaker(speakerID, rating):
    try:
        if (len(spker_map[speakerID]) <7):
            spker_map[speakerID].append(rating)
    except:
        rating_list = []
        rating_list.append(rating)
        spker_map[speakerID] = rating_list

#7 Raters (n) assign 83 speakers (N) to a total of 10 categories (1 to 10 rating)
N = 83
n = 7
k = 10

#Sum of all cells
Sum = N * n
fct = 1.0 / (N * (N-1))


#STARTs HERE
#Start from the second row -- rowID 1 (indexing starts from 0)
for row in range(1,first_sheet.nrows):
    speaker_id = int(first_sheet.cell(row,0).value)
    rating = int(first_sheet.cell(row,1).value)
    updateNumOfRaterToSpeaker(speaker_id, rating)

updated_speaker_map = dict()

#Updated speaker map has the speaker ID to its rate counter map
#Eg : Speaker 1 has 2 times 1 rating, 5 times 3 rating, etc. 
for k,v in spker_map.items():
    updated_speaker_map[k] = Counter(v)    
    

#Calculate the kappa
sum_Pj = 0
Pjs = [] #This has Pj
for k,v in updated_speaker_map.items():
    sum_ = 0
    for rating, count in v.items():
        sum_ = sum_ + (count * count)
        #print (rating,count)
    sum_ = sum_ - N
    Pj = sum_ * fct
    Pjs.append(Pj*Pj)
    sum_Pj = sum_Pj + Pj
    #print (k , Pj)
    
pis = [] #This has pi 
Number_of_Ratings = 10

for cat in range(Number_of_Ratings):
    pi = 0
    for k,v in updated_speaker_map.items():
        pi = pi + v[cat]
        
    pis.append(pi/Sum)
 
P_bar = (sum(pis)) / N
Pe_bar = sum(Pjs)

k = (P_bar - Pe_bar)/(1 - Pe_bar)

print ("Fleiss' Kappa k = ",k)
```
